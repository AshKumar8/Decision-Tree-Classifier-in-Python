Import the libraries

```
import sklearn.datasets as datasets
import pandas as pd
print("Successfully imported")
```

Loading the IRIS Dataset and creating its Dataframe

```
iris=datasets.load_iris()
print("IRIS Dataset successfully loaded\n")

df=pd.DataFrame(iris.data, columns=iris.feature_names)
print(df.head(5))

y=iris.target
print("\n\n",y)
```

Importing Decision Tree Library

```
from sklearn.tree import DecisionTreeClassifier
print("Successfully imported")
```

Defining the Decision Tree Algorithm

```
dtree=DecisionTreeClassifier()
dtree.fit(df,y)
print("Decision Tree successfully defined")
```

Importing libraries for graph Visualisation

```
from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus
print("\nSuccessfully imported")
```

Visualisation

```
def esc(code):
    return f'\033[{code}m'
print(esc('31;1;4')+'\t\t\t\t\t\t\t\tDECISION TREE VISUALISATION\n')

dot_data = StringIO()
export_graphviz(dtree,out_file=dot_data,feature_names=iris.feature_names,filled=True,rounded=False,special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```
